# Algoritma Decision Tree

```python
class DecisionTree:
    '''
        Algoritma dibuat berdasarkan tutorial dari Sebastian Mantey dengan beberapa 
        modifikasi dan penyesuaian.

        https://github.com/SebastianMantey/Decision-Tree-from-Scratch
    '''
    def __init__(self, max_depth=10, min_samples=2):
        self.min_samples = min_samples
        self.max_depth = max_depth

    def fit(self, df):
        self.df = df
        self.tree = self.__build_tree(self.df)
    
    def predict(self, test_df):
        X_test = test_df.drop(labels='labels', axis=1)
        
        self.result = []
    
        for i in range(len(X_test)):
            self.result.append(self.__predict_row(X_test.iloc[i], self.tree))

    def __predict_row(self, row, tree):
        question = list(tree.keys())[0]
        feature_name, op, value = question.split()

        if row[feature_name] <= float(value):
            answer = tree[question][0]
        else:
            answer = tree[question][1]

        if not isinstance(answer, dict):
            return answer
        else:
            residual_tree = answer
            return self.__predict_row(row, residual_tree)
        
    def __build_tree(self, df, counter=0):
        if counter == 0:
            global COLUMN_HEADERS
            COLUMN_HEADERS = df.columns

            df = df.values

        if self.__check_purity(df) or counter == self.max_depth or len(df) < self.min_samples:
            classification = self.__classify_data(df)

            return classification
        else:
            counter += 1

            potential_splits = self.__find_potential_splits(df)
            split_column, split_value = self.__find_best_split(df, potential_splits)

            lower_data, upper_data = self.__split_data(df, split_column, split_value)

            feature_name = COLUMN_HEADERS[split_column]
            question = "{} <= {}".format(feature_name, split_value)
            subtree = {question: []}

            yes_answer = self.__build_tree(lower_data, counter)
            no_answer = self.__build_tree(upper_data, counter)

            subtree[question].append(yes_answer)
            subtree[question].append(no_answer)

            return subtree

    def __check_purity(self, df):
        '''
            Return true if all classes are the same
        '''
        y = df[:, -1]

        unique_classes = np.unique(y)

        if len(unique_classes) == 1:
            return True
        else:
            return False

    def __classify_data(self, df):
        '''
            Find classes with the most count
        '''

        y = df[:, -1]

        unique_classes, unique_classes_count = np.unique(y, return_counts=True)
        max_index = unique_classes_count.argmax()

        classification = unique_classes[max_index]

        return classification

    def __find_potential_splits(self, df):
        potential_splits = {}
        _, columns_count = self.df.shape

        for c in range(columns_count - 1):
            potential_splits[c] = []
            values = df[:, c]

            unique_values = np.unique(values)

            for i in range(len(unique_values)):
                if i != 0:
                    current_value = unique_values[i]
                    previous_value = unique_values[i-1]

                    potential_split = (current_value + previous_value) / 2

                    potential_splits[c].append(potential_split)

        return potential_splits

    def __split_data(self, df, split_column, split_value):
        split_column_values = df[:, split_column]

        lower_data = df[split_column_values <= split_value]
        upper_data = df[split_column_values > split_value]

        return lower_data, upper_data

    def __calculate_entropy(self, df):
        y = df[:, -1]
        _, counts = np.unique(y, return_counts=True)

        probabilities = counts / counts.sum()

        entropy = sum(probabilities * -np.log2(probabilities))

        return entropy

    def __calculate_overall_entropy(self, lower_data, upper_data):
        data_counts = len(lower_data) + len(upper_data)

        lower_data_probability = len(lower_data) / data_counts
        upper_data_probability = len(upper_data) / data_counts

        overall_entropy = ((lower_data_probability * self.__calculate_entropy(lower_data))
                            * (upper_data_probability * self.__calculate_entropy(upper_data)))

        return overall_entropy

    def __find_best_split(self, df, potential_splits):
        overall_entropy = float("Inf")

        for c in potential_splits:
            for v in potential_splits[c]:
                lower_data, upper_data = self.__split_data(df, c, v)
                current_overall_entropy = self.__calculate_overall_entropy(lower_data, upper_data)

            if current_overall_entropy < overall_entropy:
                overall_entropy = current_overall_entropy
                best_split_column = c
                best_split_value = v

        return best_split_column, best_split_value
```

# Pembuatan Model dari Data

```python
start_training_time = time.time()

tr = DecisionTree()
tr.fit(df=train_df)

elapsed = time.time() - start_training_time
print(f"Finished training with duration of {int(elapsed/60)} minutes {round(float(elapsed%60), 3)} seconds.")
```

# Membuat Prediksi dan Generate Report

```python
from sklearn.metrics import classification_report

print(classification_report(tr.result, y_test))
```
